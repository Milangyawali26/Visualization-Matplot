```python
#  sactter plot is used to identify whether there is co relation between variables and values

import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

```

```python
data=pd.read_csv("data_scatter.csv")
data.head(3)
```

```python
likes=data['likes']
view_count=data['view_count']
ratio=data['ratio']
```

```python
plt.scatter(view_count,likes,
            c=ratio,cmap='summer',
            edgecolors='black',
            linewidths=1,alpha=0.5)

plt.xscale('log')
plt.yscale('log')    #this helps to overcome problem from outliers


cbar=plt.colorbar()
cbar.set_label("like/ dislike ratio ")

plt.title('trending youtube videos')
plt.xlabel('view count')
plt.ylabel('total likes')

plt.tight_layout()
```

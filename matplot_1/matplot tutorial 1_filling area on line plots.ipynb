```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('data.csv')
data.head()
```

```python
overall_median=data.describe().median().median()
print(overall_median)
```

```python
ages=data['Age']
dev_salaries=data['All_Devs']
py_salaries=data['Python']
js_salaries=data['JavaScript']

plt.plot(ages,dev_salaries,linestyle='--', label='all devs')

plt.plot(ages,py_salaries,label='python')

plt.title("median salary (usd) by ages")
plt.xlabel("ages")
plt.ylabel("salaries")
plt.tight_layout()
plt.grid("white")
plt.legend()

#  fill between the line s

plt.fill_between(ages,py_salaries,overall_median,
                 where=(py_salaries>overall_median),
                 alpha=0.25,  )   # alpha is used to define the transparency

plt.fill_between(ages,py_salaries,overall_median,
                 where=(py_salaries<=overall_median),
                 alpha=0.25,  )   # alpha is used to define the transparency
```

```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")


slices=[6,4,3,2]
labels=['six','four','three','two']
explode=[0,0,0,0.1]

plt.pie(slices,labels=labels,
        shadow=True,
        
        explode=explode,    # explode le tyo tukra lai bahira nikalxa
        wedgeprops={'edgecolor':'black'})   # wedgeprops le edge lai black banairako xa


plt.title("my awesome pie chart ")
plt.tight_layout()
plt.show()
```
